```
import random

def kromosom():
    kromosom = []
    for i in range(6):
        kromosom.append(random.randint(0,9))

    return kromosom
```

```
def populasi():
    populasi = []
    for i in range(8):
        populasi.append(kromosom())

    return populasi
```

```
def split(num):
    s = len(num)//2

    return num[:s], num[s:]
```

```
def decodechrom(num, minimum, maximum):
    atas = 0
    bawah = 0
    for i in range(len(num)):
        cntr = num[i]
        atas += (cntr*(10**-(i+1)))
        bawah += (9*(10**-(i+1)))
    decode = minimum + (((maximum - minimum)*atas)/bawah)

    return decode
```

```
import math

def rumus(x1, x2):
    persamaan = (math.cos(x1)*math.sin(x2))-(x1/(x2**2+1))

    return persamaan
```

```
def fitness(h):
    f = 1/(h+0.1)

    return f
```

```
def ortu(pop, fit, totfit):
  ran = random.random()
  i = 0
  while(ran > 0):
    ran -= fit[i]/totfit
    i += 1
    if (i == len(pop)-1):
      break
  parent = pop[i]

  return parent
```

```
def crossover(parentA, parentB):
  crossA = []
  crossB = []
  crs = []
  prob = random.random()
  if(prob < 0.8):
    pointer = random.randint(0,7)
    crossA[:pointer] = parentA[:pointer]
    crossA[pointer:] = parentB[pointer:]
    crossB[:pointer] = parentB[:pointer]
    crossB[pointer:] = parentA[pointer:]
    crs.append(crossA)
    crs.append(crossB)
  else:
    crs.append(parentA)
    crs.append(parentB)

  return crs
```

```
def mutasi(crossA, crossB):
  prob = random.random()
  if (prob < 0.2):
    for i in range(len(crossA)):
      rand = random.random()
      if (rand < 0.1):
        crossA[i] = random.randint(0,9)
    for i in range(len(crossB)):
      rand = random.random()
      if (rand < 0.1):
        crossB[i] = random.randint(0,9)
  crs = []
  crs.append(crossA)
  crs.append(crossB)

  return crs
```

```
def terbaik(pop):
  maxFit = -9999
  p = []

  for i in range(len(pop)):
    p = pop[i]
    s1, s2 = split(p)
    x1 = decodechrom(s1, -1, 2)
    x2 = decodechrom(s2, -1, 1)
    r = rumus(x1, x2)
    fit = fitness(r)

    if (fit > maxFit):
      maxFit = fit
      maxP = p

  return maxP
```
